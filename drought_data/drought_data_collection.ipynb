```python
# Dependencies
import requests
import json
import pandas as pd
```

```python
# Drought Monitor API specification can be found at:
# https://droughtmonitor.unl.edu/DmData/DataDownload/WebServiceInfo.aspx#comp

# Configure API parameters
url_stub = "https://usdmdataservices.unl.edu/api/CountyStatistics/GetDroughtSeverityStatisticsByPopulationPercent?"
headers = {"Accept": "application/json",}
# cities = [('Chicago','17031'), ('Boston', '25025'), ('Baltimore', '24510'), ('New Orleans', '22071')]

# URL constructor function
def req_const(city):
    url = url_stub + '&aoi=' + str(city) + '&startdate=1/1/2000&enddate=12/31/2014&statisticsType=2'
    return url
```

```python
# Retrieve Chicago data
req = requests.get(req_const('17031'), headers=headers) 
res = req.json()
chicago_data = pd.json_normalize(res)
```

```python
# Retrieve Boston data
req = requests.get(req_const('25025'), headers=headers) 
res = req.json()
boston_data = pd.json_normalize(res)
```

```python
# Retrieve Baltimore data
req = requests.get(req_const('24510'), headers=headers) 
res = req.json()
baltimore_data = pd.json_normalize(res)
```

```python
# Retrieve New Orleans data
req = requests.get(req_const('22071'), headers=headers) 
res = req.json()
nola_data = pd.json_normalize(res)
```

```python
# Create combined dataframe
drought_data = pd.concat([chicago_data, boston_data, baltimore_data, nola_data], axis=0, join='outer', ignore_index=True)
```

```python
drought_data
```

```python
# Remove extraneous Columns
drought_data.drop(columns=['MapDate', 'FIPS', 'StatisticFormatID'], inplace=True)
```

```python
# Convert column types
drought_data[['None', 'D0', 'D1', 'D2', 'D3', 'D4']] = drought_data[['None', 'D0', 'D1', 'D2', 'D3', 'D4']].apply(pd.to_numeric)
drought_data[['ValidStart', 'ValidEnd']] = drought_data[['ValidStart', 'ValidEnd']].apply(pd.to_datetime)
```

```python
# Group data into monthly records
drought_data['Year'] = drought_data.ValidStart.dt.year
drought_data['Month'] = drought_data.ValidStart.dt.month
drought_data = drought_data.groupby(['County', 'Year', 'Month']).mean().reset_index()
```

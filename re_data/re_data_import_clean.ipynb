```python
import pandas as pd
```

```python
# read csv data
cols = ['Region', 'Period End', 'New Construction?', 'Homes Sold', 'Homes Sold YoY', '% of Homes Sold Is New Construction', 'Inventory', 'Inventory YoY', '% of Inventory Is New Construction', 'New Listings', 'New Listings YoY', 'Median Sale Price', 'Median Sale Price YoY', 'Median Ppsf', 'Median Ppsf YoY', 'Median List Price', 'Avg Sale To List', 'Sold Above List', 'Months Of Supply']
re_raw = pd.read_csv('data/Redfin RE Data.csv', sep=",|\t", header=1, encoding = 'unicode_escape', engine='python')
```

```python
re_raw
```

```python
# Convert column types
re_data = pd.DataFrame()
re_data[['Homes_Sold', 'Homes_Sold_YoY', '%_Sold_Is_New_Construction', 'Inventory', 'Inventory_YoY', '%_Inventory_Is_New_Construction', 'New_Listings', 'New_Listings_YoY', 'Median_Sale_Price', 'Median_Sale_Price_YoY', 'Median_Ppsf', 'Median_Ppsf_YoY', 'Median_List_Price', 'Avg_Sale_To_List', 'Sold_Above_List', 'Months_Of_Supply']] = re_raw[['2', '830', '27.5', '93.52', '11', '541', '220', '000', '-1.8', '142', '-1.1', '195', '900', '96.7', '16.0', '4']].apply(pd.to_numeric)
re_data[['County', 'Existing']] = re_raw[['Baltimore', 'Existing']]
re_data[['End_Date']] = re_raw[['12/31/2014']].apply(pd.to_datetime)
```

```python
for col in re_data.columns:
    print(col)
```

```python
# Rename the cities
re_data['County'] = re_data['County'].replace('Baltimore', 'Baltimore City')
re_data['County'] = re_data['County'].replace('New Orleans', 'Orleans Parish')
re_data['County'] = re_data['County'].replace('Boston', 'Suffolk County')
re_data['County'] = re_data['County'].replace('Chicago', 'Cook County')
```

```python
# Group data into monthly records
re_data['Year'] = re_data.End_Date.dt.year
re_data['Month'] = re_data.End_Date.dt.month
re_data = re_data.groupby(['County', 'Year', 'Month', 'Existing']).mean().reset_index()
```

```python
# export formatted data
re_data.to_pickle('data/re_data.pk1')
```

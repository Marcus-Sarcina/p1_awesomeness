```python
import pandas as pd
import seaborn as sns
```

```python
file_path = 'murder data.csv'
murder_data = pd.read_csv(file_path)
murder_data['date'] = pd.to_datetime(murder_data['Year'].astype(str) + '-' + murder_data['Month'].astype(str))
murder_data.drop(columns=['Record ID', 'Agency Code', 'Agency Name', 'Agency Type', 'Victim Race', 'Victim Ethnicity', 'Perpetrator Sex', 'Perpetrator Age', 'Perpetrator Race', 'Perpetrator Ethnicity', 'Relationship', 'Record Source', 'Victim Sex', 'Victim Age', 'Perpetrator Count'], inplace=True)
print(murder_data.head())
```

```python
Chi_data = murder_data[(murder_data["City"] == "Cook") &
                      (murder_data["State"] == "Illinois") &
                      (murder_data["Year"] > 1999) &
                      (murder_data["Year"] < 2006)
                     ]
Chi_data

Nola_data = murder_data[(murder_data["City"] == "Orleans") &
                      (murder_data["State"] == "Louisiana") &
                      (murder_data["Year"] > 1999) &
                      (murder_data["Year"] < 2006)
                     ]
Nola_data

Boston_data = murder_data[(murder_data["City"] == "Suffolk") &
                      (murder_data["State"] == "Massachusetts") &
                      (murder_data["Year"] > 1999) &
                      (murder_data["Year"] < 2006)
                     ]
Boston_data

Balt_data = murder_data[(murder_data["City"] == "Baltimore city") &
                      (murder_data["State"] == "Maryland") &
                      (murder_data["Year"] > 1999) &
                      (murder_data["Year"] < 2006)
                     ]
```

```python
print(Chi_data, Nola_data)
```

```python
Chi_data.plot('date', 'Victim Count')
```
